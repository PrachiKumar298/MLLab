Iris

```python
from sklearn.datasets import load_iris
import pandas as pd

# Load the Iris dataset
iris = load_iris()

# Create a DataFrame for better readability
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target
iris_df['target_names'] = iris.target_names[iris.target]

# Display the first 5 rows of the dataset
display(iris_df.head())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Separate features (X) and target (y)
X = iris_df.drop(columns=['target', 'target_names'])
y = iris_df['target']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print(f"Shape of X_train: {X_train.shape}")
print(f"Shape of X_test: {X_test.shape}")
print(f"Shape of y_train: {y_train.shape}")
print(f"Shape of y_test: {y_test.shape}")
```

```python
# Initialize StandardScaler
scaler = StandardScaler()

# Fit the scaler on the training data and transform both training and testing data
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Display the first 5 rows of the scaled training data
display(pd.DataFrame(X_train_scaled, columns=X.columns).head())
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, classification_report

# Initialize the MLP Classifier
mlp = MLPClassifier(hidden_layer_sizes=(10, 8), max_iter=1000)

# Train the classifier
mlp.fit(X_train_scaled, y_train)

# Make predictions on the scaled test data
y_pred = mlp.predict(X_test_scaled)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

Breast Cancer

```python
from sklearn.datasets import load_breast_cancer
import pandas as pd

# Load the Breast Cancer dataset
breast_cancer = load_breast_cancer()

# Create a DataFrame for better readability
bc_df = pd.DataFrame(data=breast_cancer.data, columns=breast_cancer.feature_names)
bc_df['target'] = breast_cancer.target
bc_df['target_names'] = breast_cancer.target_names[breast_cancer.target]

# Display the first 5 rows of the dataset
display(bc_df.head())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Separate features (X_bc) and target (y_bc)
X_bc = bc_df.drop(columns=['target', 'target_names'])
y_bc = bc_df['target']

# Split the data into training and testing sets
X_bc_train, X_bc_test, y_bc_train, y_bc_test = train_test_split(X_bc, y_bc, test_size=0.3, random_state=42)

print(f"Shape of X_bc_train: {X_bc_train.shape}")
print(f"Shape of X_bc_test: {X_bc_test.shape}")
print(f"Shape of y_bc_train: {y_bc_train.shape}")
print(f"Shape of y_bc_test: {y_bc_test.shape}")
```

```python
# Initialize StandardScaler for Breast Cancer dataset
scaler_bc = StandardScaler()

# Fit the scaler on the training data and transform both training and testing data
X_bc_train_scaled = scaler_bc.fit_transform(X_bc_train)
X_bc_test_scaled = scaler_bc.transform(X_bc_test)

# Display the first 5 rows of the scaled training data
display(pd.DataFrame(X_bc_train_scaled, columns=X_bc.columns).head())
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, classification_report

# Initialize the MLP Classifier with similar parameters
mlp_bc = MLPClassifier(hidden_layer_sizes=(10, 8), max_iter=1000, random_state=42)

# Train the classifier on the scaled Breast Cancer training data
mlp_bc.fit(X_bc_train_scaled, y_bc_train)

# Make predictions on the scaled Breast Cancer test data
y_bc_pred = mlp_bc.predict(X_bc_test_scaled)

# Evaluate the model
accuracy_bc = accuracy_score(y_bc_test, y_bc_pred)
print(f"Breast Cancer MLP Accuracy: {accuracy_bc:.4f}")
print("\nBreast Cancer Classification Report:")
print(classification_report(y_bc_test, y_bc_pred))
```

```python
import pickle
import pandas as pd
import numpy as np

# Load the hf_seq_data_array.pkl file
with open('DeepHF/data/hf_seq_data_array.pkl', 'rb') as f:
    hf_data = pickle.load(f)

print(f"Type of hf_data: {type(hf_data)}")
print(f"Length of hf_data tuple: {len(hf_data)}")

print("\nInspecting each element in the tuple:")
for i, item in enumerate(hf_data):
    print(f"\nElement {i}:")
    print(f"  Type: {type(item)}")
    if hasattr(item, 'shape'):
        print(f"  Shape: {item.shape}")
    else:
        print(f"  Length: {len(item)}")

    # Display first few elements if it's an array or list-like
    if isinstance(item, np.ndarray) or isinstance(item, list):
        print("  First 5 elements:")
        try:
            display(pd.DataFrame(item).head())
        except (TypeError, ValueError):
            print(item[:5])
    elif isinstance(item, pd.DataFrame):
        display(item.head())
    else:
        print(item)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPRegressor # Changed from MLPClassifier
from sklearn.metrics import mean_squared_error, r2_score # Changed metrics for regression
import numpy as np

# Assuming hf_data[0] and hf_data[1] are features and hf_data[2] is the target
# Concatenate the feature arrays horizontally
X_hf = np.hstack((hf_data[0], hf_data[1]))
y_hf = hf_data[2]

print(f"Shape of combined features (X_hf): {X_hf.shape}")
print(f"Shape of target (y_hf): {y_hf.shape}")

# Split the data into training and testing sets
# Removed stratify=y_hf because y_hf is a continuous variable.
X_hf_train, X_hf_test, y_hf_train, y_hf_test = train_test_split(X_hf, y_hf, test_size=0.3, random_state=42)

print(f"\nShape of X_hf_train: {X_hf_train.shape}")
print(f"Shape of X_hf_test: {X_hf_test.shape}")
print(f"Shape of y_hf_train: {y_hf_train.shape}")
print(f"Shape of y_hf_test: {y_hf_test.shape}")

# Initialize StandardScaler for the DeepHF dataset
scaler_hf = StandardScaler()

# Fit the scaler on the training data and transform both training and testing data
X_hf_train_scaled = scaler_hf.fit_transform(X_hf_train)
X_hf_test_scaled = scaler_hf.transform(X_hf_test)

# Initialize the MLP Regressor
# Using similar parameters as previous examples, but adjusting hidden_layer_sizes if needed
mlp_hf = MLPRegressor(hidden_layer_sizes=(50, 25), max_iter=1000, random_state=42, verbose=True)

# Train the regressor on the scaled DeepHF training data
mlp_hf.fit(X_hf_train_scaled, y_hf_train)

# Make predictions on the scaled DeepHF test data
y_hf_pred = mlp_hf.predict(X_hf_test_scaled)

# Evaluate the model using regression metrics
mse = mean_squared_error(y_hf_test, y_hf_pred)
r2 = r2_score(y_hf_test, y_hf_pred)
print(f"\nDeepHF MLP Mean Squared Error: {mse:.4f}")
print(f"DeepHF MLP R2 Score: {r2:.4f}")
```
