Iris

```python
from sklearn.datasets import load_iris

iris = load_iris()
X = iris.data
y = iris.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print(f"X_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"y_test shape: {y_test.shape}")
```

```python
import pandas as pd
import numpy as np


from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier

from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import StackingClassifier

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    classification_report,
    confusion_matrix
)
```

```python
bagging_model = BaggingClassifier(
    estimator=DecisionTreeClassifier(),
    n_estimators=100,
    random_state=42
)
```

```python
adaboost_model = AdaBoostClassifier(
    estimator=DecisionTreeClassifier(max_depth=1),
    n_estimators=100,
    learning_rate=0.5,
    random_state=42
)

adaboost_model.fit(X_train, y_train)
y_pred_adaboost = adaboost_model.predict(X_test)

gboost_model = GradientBoostingClassifier(
    n_estimators=100,
    learning_rate=0.1,
    random_state=42
)
gboost_model.fit(X_train, y_train)
y_pred_gboost = gboost_model.predict(X_test)
```

```python
base_learners = [
    ('dt', DecisionTreeClassifier()),
    ('knn', KNeighborsClassifier()),
    ('lr', LogisticRegression(max_iter=1000))
]

stacking_model = StackingClassifier(
    estimators=base_learners,
    final_estimator=LogisticRegression(),
    cv=5
)
```

```python
print('AdaBoost Classifier Evaluation:')
print(f'Accuracy: {accuracy_score(y_test, y_pred_adaboost):.4f}')
print(f'Precision: {precision_score(y_test, y_pred_adaboost, average="weighted"):.4f}')
print(f'Recall: {recall_score(y_test, y_pred_adaboost, average="weighted"):.4f}')
print(f'F1-score: {f1_score(y_test, y_pred_adaboost, average="weighted"):.4f}')
print('\nClassification Report:')
print(classification_report(y_test, y_pred_adaboost))
print('\nConfusion Matrix:')
print(confusion_matrix(y_test, y_pred_adaboost))
```

```python
print('Gradient Boosting Classifier Evaluation:')
print(f'Accuracy: {accuracy_score(y_test, y_pred_gboost):.4f}')
print(f'Precision: {precision_score(y_test, y_pred_gboost, average="weighted"):.4f}')
print(f'Recall: {recall_score(y_test, y_pred_gboost, average="weighted"):.4f}')
print(f'F1-score: {f1_score(y_test, y_pred_gboost, average="weighted"):.4f}')
print('\nClassification Report:')
print(classification_report(y_test, y_pred_gboost))
print('\nConfusion Matrix:')
print(confusion_matrix(y_test, y_pred_gboost))
```

```python
bagging_model.fit(X_train, y_train)
y_pred_bagging = bagging_model.predict(X_test)

print('Bagging Classifier Evaluation:')
print(f'Accuracy: {accuracy_score(y_test, y_pred_bagging):.4f}')
print(f'Precision: {precision_score(y_test, y_pred_bagging, average="weighted"):.4f}')
print(f'Recall: {recall_score(y_test, y_pred_bagging, average="weighted"):.4f}')
print(f'F1-score: {f1_score(y_test, y_pred_bagging, average="weighted"):.4f}')
print('\nClassification Report:')
print(classification_report(y_test, y_pred_bagging))
print('\nConfusion Matrix:')
print(confusion_matrix(y_test, y_pred_bagging))
```

```python
stacking_model.fit(X_train, y_train)
y_pred_stacking = stacking_model.predict(X_test)

print('Stacking Classifier Evaluation:')
print(f'Accuracy: {accuracy_score(y_test, y_pred_stacking):.4f}')
print(f'Precision: {precision_score(y_test, y_pred_stacking, average="weighted"):.4f}')
print(f'Recall: {recall_score(y_test, y_pred_stacking, average="weighted"):.4f}')
print(f'F1-score: {f1_score(y_test, y_pred_stacking, average="weighted"):.4f}')
print('\nClassification Report:')
print(classification_report(y_test, y_pred_stacking))
print('\nConfusion Matrix:')
print(confusion_matrix(y_test, y_pred_stacking))
```

CRISPR

```python
import pandas as pd

df_crispr = pd.read_csv('/content/crispr_dataset.csv')

print('Crispr Dataset Info:')
df_crispr.info()
print('\nFirst 5 rows of Crispr Dataset:')
display(df_crispr.head())
```

```python
def count_nucleotides(sequence):
    return {
        'A_count': sequence.count('A'),
        'C_count': sequence.count('C'),
        'G_count': sequence.count('G'),
        'T_count': sequence.count('T')
    }

nucleotide_features = df_crispr['gRNA'].apply(count_nucleotides).apply(pd.Series)
df_crispr_processed = pd.concat([df_crispr, nucleotide_features], axis=1)

X_crispr = df_crispr_processed[['A_count', 'C_count', 'G_count', 'T_count']]
y_crispr = df_crispr_processed['efficiency']

X_train_crispr, X_test_crispr, y_train_crispr, y_test_crispr = train_test_split(X_crispr, y_crispr, test_size=0.3, random_state=42)

print(f"X_train_crispr shape: {X_train_crispr.shape}")
print(f"X_test_crispr shape: {X_test_crispr.shape}")
print(f"y_train_crispr shape: {y_train_crispr.shape}")
print(f"y_test_crispr shape: {y_test_crispr.shape}")

display(X_crispr.head())
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor

from sklearn.ensemble import BaggingRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import StackingRegressor

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

print("--- Bagging Regressor ---")
bagging_model_crispr = BaggingRegressor(
    estimator=DecisionTreeRegressor(),
    n_estimators=100,
    random_state=42
)
bagging_model_crispr.fit(X_train_crispr, y_train_crispr)
y_pred_bagging_crispr = bagging_model_crispr.predict(X_test_crispr)

print(f'MAE: {mean_absolute_error(y_test_crispr, y_pred_bagging_crispr):.4f}')
print(f'MSE: {mean_squared_error(y_test_crispr, y_pred_bagging_crispr):.4f}')
print(f'R2 Score: {r2_score(y_test_crispr, y_pred_bagging_crispr):.4f}')

print("\n--- AdaBoost Regressor ---")
adaboost_model_crispr = AdaBoostRegressor(
    estimator=DecisionTreeRegressor(max_depth=4),
    n_estimators=100,
    learning_rate=0.5,
    random_state=42
)
adaboost_model_crispr.fit(X_train_crispr, y_train_crispr)
y_pred_adaboost_crispr = adaboost_model_crispr.predict(X_test_crispr)

print(f'MAE: {mean_absolute_error(y_test_crispr, y_pred_adaboost_crispr):.4f}')
print(f'MSE: {mean_squared_error(y_test_crispr, y_pred_adaboost_crispr):.4f}')
print(f'R2 Score: {r2_score(y_test_crispr, y_pred_adaboost_crispr):.4f}')

print("\n--- Gradient Boosting Regressor ---")
gboost_model_crispr = GradientBoostingRegressor(
    n_estimators=100,
    learning_rate=0.1,
    random_state=42
)
gboost_model_crispr.fit(X_train_crispr, y_train_crispr)
y_pred_gboost_crispr = gboost_model_crispr.predict(X_test_crispr)

print(f'MAE: {mean_absolute_error(y_test_crispr, y_pred_gboost_crispr):.4f}')
print(f'MSE: {mean_squared_error(y_test_crispr, y_pred_gboost_crispr):.4f}')
print(f'R2 Score: {r2_score(y_test_crispr, y_pred_gboost_crispr):.4f}')

print("\n--- Stacking Regressor ---")
base_learners_crispr = [
    ('dt', DecisionTreeRegressor()),
    ('knn', KNeighborsRegressor()),
    ('lr', LinearRegression())
]

stacking_model_crispr = StackingRegressor(
    estimators=base_learners_crispr,
    final_estimator=LinearRegression(),
    cv=5
)
stacking_model_crispr.fit(X_train_crispr, y_train_crispr)
y_pred_stacking_crispr = stacking_model_crispr.predict(X_test_crispr)

print(f'MAE: {mean_absolute_error(y_test_crispr, y_pred_stacking_crispr):.4f}')
print(f'MSE: {mean_squared_error(y_test_crispr, y_pred_stacking_crispr):.4f}')
print(f'R2 Score: {r2_score(y_test_crispr, y_pred_stacking_crispr):.4f}')
```
